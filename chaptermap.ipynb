```python
import json
import pandas as pd

# Load JSON data from a file
json_file_path = 'genai.chapter.json'  # Update this path if necessary

with open(json_file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Convert JSON data to a Pandas DataFrame
df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

# Bar chart for verse counts in each chapter
plt.figure(figsize=(10, 6))
plt.bar(df['chapterName'], df['verseCount'], color='skyblue')
plt.title('Verse Count per Chapter')
plt.xlabel('Chapter Name')
plt.ylabel('Number of Verses')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Pie chart for chapters by language
language_counts = df['language'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(language_counts, labels=language_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Chapters by Language')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Load the JSON data
with open('genai.chapter.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Extract chapter names and verse counts
chapter_names = [chapter['chapterName'] for chapter in data]
verse_counts = [chapter['verseCount'] for chapter in data]

# Number of chapters
num_chapters = len(chapter_names)

# Create an array of angles for each chapter
angles = np.linspace(0, 2 * np.pi, num_chapters, endpoint=False).tolist()

# The bars need to be "closed" for a complete circular graph
verse_counts += verse_counts[:1]
angles += angles[:1]

# Create the figure and polar axis
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': 'polar'})

# Draw the bars
bars = ax.bar(angles[:-1], verse_counts[:-1], color='skyblue', alpha=0.7, edgecolor='black')

# Add labels to each bar
for i, bar in enumerate(bars):
    angle = bar.get_x() + bar.get_width() / 2  # Get the angle for the label
    ax.text(angle, bar.get_height() + 1, chapter_names[i], ha='center', va='bottom', fontsize=10)

# Set the title and customize the grid
ax.set_title('Number of Verses per Chapter (Radial Chart)', fontsize=16)
ax.set_xticks(angles[:-1])  # Set the ticks to be at the angles of each chapter
ax.set_xticklabels(chapter_names)  # Set the labels to be the chapter names

# Show grid lines
ax.grid(True)

# Display the chart
plt.tight_layout()
plt.show()
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Load the JSON data
with open('genai.chapter.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Extract chapter names and verse counts
chapter_names = [chapter['chapterName'] for chapter in data]
verse_counts = [chapter['verseCount'] for chapter in data]

# Number of chapters
num_chapters = len(chapter_names)

# Create an array of angles for each chapter
angles = np.linspace(0, 2 * np.pi, num_chapters, endpoint=False).tolist()

# The bars need to be "closed" for a complete circular graph
verse_counts += verse_counts[:1]
angles += angles[:1]

# Create the figure and polar axis
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': 'polar'})

# Draw the bars
bars = ax.bar(angles[:-1], verse_counts[:-1], color='skyblue', alpha=0.7, edgecolor='black')

# Add labels to each bar with proper Unicode support
for i, bar in enumerate(bars):
    angle = bar.get_x() + bar.get_width() / 2  # Get the angle for the label
    ax.text(angle, bar.get_height() + 1, chapter_names[i], ha='center', va='bottom', fontsize=10)

# Set the title and customize the grid
ax.set_title('Number of Verses per Chapter (Radial Chart)', fontsize=16)
ax.set_xticks(angles[:-1])  # Set the ticks to be at the angles of each chapter
ax.set_xticklabels(chapter_names)  # Set the labels to be the chapter names

# Show grid lines
ax.grid(True)

# Adjust font properties to support multiple languages (including Telugu and Hindi)
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial Unicode MS']  # Use a font that supports Unicode characters

# Display the chart
plt.tight_layout()
plt.show()
```

```python
# Step 1: Install Necessary Libraries (if not already installed)
!pip install pandas matplotlib plotly

# Step 2: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import json

# Step 3: Load the JSON File
file_path = '/content/genai.chapter.json'  # Update this to the actual file path
with open(file_path) as f:
    data = json.load(f)

# Step 4: Create a DataFrame from JSON
df = pd.json_normalize(data)

# Step 5: Basic Analysis
# Preview the data
print(df.head())

# Step 6: Bar Chart - Number of Verses per Chapter
plt.figure(figsize=(10,6))
df.plot(x='chapterName', y='verseCount', kind='bar', legend=False)
plt.title('Number of Verses per Chapter')
plt.ylabel('Verse Count')
plt.xticks(rotation=90)
plt.show()

# Step 7: Interactive Pie Chart using Plotly - Verse Count Distribution
fig = px.pie(df, names='chapterName', values='verseCount', title='Verse Distribution Across Chapters')
fig.show()

# Step 8: Line Graph - Verses per Chapter Number
plt.figure(figsize=(10,6))
plt.plot(df['chapterNumber'], df['verseCount'], marker='o')
plt.title('Verses Across Chapters')
plt.xlabel('Chapter Number')
plt.ylabel('Verse Count')
plt.grid(True)
plt.show()

# Step 9: Treemap - Summary of Each Chapter
fig = px.treemap(df, path=['chapterName'], values='verseCount', title="Treemap of Chapter Summary")
fig.show()
```
